```python
import pandas as pd

def is_prime(n):
    """Checks if a number n is prime."""
    if n <= 1: return False
    if n <= 3: return True
    if n % 2 == 0 or n % 3 == 0: return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def get_nth_prime(n):
    """Returns the n-th prime number (1-based index)."""

    if n < 1: return None
    count = 0
    candidate = 2
    while True:
        if is_prime(candidate):
            count += 1
            if count == n:
                return candidate
        candidate += 1

df_raw = pd.read_excel(quran_url)

verses_count = []
current_count = 0
for n in df_raw['Numbers']:
    if n != 0:
        current_count += 1
    else:
        if current_count > 0:
            verses_count.append(current_count)
            current_count = 0
if current_count > 0:
    verses_count.append(current_count)

df = pd.DataFrame({
    'Surah_Number': range(1, len(verses_count) + 1),
    'Verses': verses_count
})

df['Is_Verse_Prime'] = df['Verses'].apply(is_prime)

def calculate_symmetry_value(row):
    v = row['Verses']
    if row['Is_Verse_Prime']:
        return get_nth_prime(v)
    else:
        return v

df['Calculated_Value'] = df.apply(calculate_symmetry_value, axis=1)

prime_group_sum = df[df['Is_Verse_Prime'] == True]['Calculated_Value'].sum()
non_prime_group_sum = df[df['Is_Verse_Prime'] == False]['Calculated_Value'].sum()

print("="*60)
print("QURAN NUMERICAL SYMMETRY VERIFICATION")
print("="*60)

print(f"Total Surahs: {len(df)}")

print(f"\n[GROUP 1] Surahs with PRIME Verse Counts")
print(f"Count: {len(df[df['Is_Verse_Prime'] == True])}")
print("Formula: Sum of the N-th Prime Number (where N is the verse count)")
print(f"--> Result Sum: {prime_group_sum}")

print("-" * 60)

print(f"\n[GROUP 2] Surahs with NON-PRIME Verse Counts")
print(f"Count: {len(df[df['Is_Verse_Prime'] == False])}")
print("Formula: Sum of the Verse Counts themselves")
print(f"--> Result Sum: {non_prime_group_sum}")

print("="*60)
print(f"FINAL VERIFICATION: {prime_group_sum} == {non_prime_group_sum} ?")
print(f"MATCH: {'YES' if prime_group_sum == non_prime_group_sum else 'NO'}")
print("="*60)
```

# The Golden Ratio in the Quran
A divine constant hidden within repeating and non-repeating sums

---

> After calculating the sum of each Surah's chapter number and its verse count, we obtain **114 results**, one for each Surah. We then categorize these sums into **repeating** (occur more than once) and **non-repeating** (occur only once) values.

---

### Sum Categorization

| **7906** | **4885** |
| :---: | :---: |
| **Repeating Sums** | **Non-Repeating Sums** |
| Occur more than once | Occur only once |

### The Ratio

| Repeating | | Non-Repeating | | **The Golden Ratio** |
| :---: | :---: | :---: | :---: | :---: |
| <code style="font-size: 18px; color: #ffffff; background-color: #00897b; padding: 5px; border-radius: 3px;">**7906**</code> | **/** | <code style="font-size: 18px; color: #ffffff; background-color: #00897b; padding: 5px; border-radius: 3px;">**4885**</code> | **=** | <code style="font-size: 22px; color: #ffffff; background-color: #f57c00; padding: 8px; border-radius: 5px;">**1.618**</code> |

---

> The ratio of repeating to non-repeating sums corresponds precisely to the **Golden Ratio (1.618...)**, a remarkable mathematical constant that appears throughout nature, art, and architecture. Its presence within the Quran's numerical structure adds yet another layer to its extraordinary mathematical design.

```python
import pandas as pd
from collections import Counter

df = pd.read_excel(quran_url)

surah_verse_counts = []
current_count = 0
numbers = df['Numbers'].tolist()

for num in numbers:
    if num == 0:
        if current_count > 0:
            surah_verse_counts.append(current_count)
        current_count = 0
    else:
        current_count += 1

if current_count > 0:
    surah_verse_counts.append(current_count)

surah_sums = []
for i, verse_count in enumerate(surah_verse_counts):
    surah_num = i + 1
    total_val = surah_num + verse_count
    surah_sums.append(total_val)

counts = Counter(surah_sums)

repetitive_sums = [val for val in surah_sums if counts[val] > 1]
non_repetitive_sums = [val for val in surah_sums if counts[val] == 1]

num_repetitive = len(repetitive_sums)
num_non_repetitive = len(non_repetitive_sums)
sum_repetitive = sum(repetitive_sums)
sum_non_repetitive = sum(non_repetitive_sums)

print(f"Total Surahs: {len(surah_sums)}")
print("-" * 30)
print(f"Number of Repetitive Surahs: {num_repetitive}")
print(f"Number of Non-Repetitive Surahs: {num_non_repetitive}")
print("-" * 30)
print(f"Sum of Repetitive Numbers: {sum_repetitive}")
print(f"Sum of Non-Repetitive Numbers: {sum_non_repetitive}")
print("-" * 30)
```
