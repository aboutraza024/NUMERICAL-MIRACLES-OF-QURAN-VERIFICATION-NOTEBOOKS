```python
import pandas as pd

def get_unique_letters(text):
    char_map = {
        'أ': 'ا', 'ا': 'ا', 'إ': 'ا', 'ء': 'ا', 'ئ': 'ا', 'ؤ': 'ا', 'آ': 'ا',
        'ة': 'ه', 'ه': 'ه',
        'ي': 'ي', 'ى': 'ي'
    }

    text = text.replace(" ", "")

    unique_chars = set()
    for char in text:
        normalized_char = char_map.get(char, char)
        unique_chars.add(normalized_char)

    return unique_chars, len(unique_chars)

df = pd.read_excel(quran_url)

kawthar_idx = df[df['Verses'].str.contains('الكوثر', na=False)].index[0]

verses = df.iloc[kawthar_idx : kawthar_idx+3]['Verses'].tolist()

for i, verse in enumerate(verses, 1):
    unique_chars, count = get_unique_letters(verse)
    print(f"Verse {i}: {verse}")
    print(f"Total Unique Letters: {count}")
    print(f"Unique Letters Set: {', '.join(unique_chars)}")
    print("-" * 50)
```

```python
import pandas as pd

char_map = {
    'أ': 'ا', 'إ': 'ا', 'ء': 'ا', 'ئ': 'ا', 'ؤ': 'ا', 'آ': 'ا',
    'ة': 'ه',
    'ى': 'ي'
}

def apply_char_map(text):
    for k, v in char_map.items():
        text = text.replace(k, v)
    return text


try:
    df = pd.read_excel(quran_url, names=["Verses", "Numbers"], header=0)

    kawthar_matches = df[df['Verses'].str.contains('الكوثر', na=False)]
    if kawthar_matches.empty:
        raise ValueError("Surah Al-Kawthar nahi mili. Text check karein.")

    kawthar_idx = kawthar_matches.index[0]
    surah_kawthar_verses = df['Verses'].iloc[kawthar_idx:kawthar_idx+3].tolist()

    all_ending_words_str = ""
    all_second_words_str = ""
    all_remaining_words_list = []

    for verse in surah_kawthar_verses:
        words = verse.split()
        if len(words) == 0:
            continue

        last_word = words[-1]

        if len(words) >= 3:
            second_word = words[1]
            remaining = [words[0]] + words[2:-1]
        elif len(words) == 2:
            second_word = words[1]
            remaining = [words[0]]
        else:
            second_word = ""
            remaining = []

        all_ending_words_str += last_word
        all_second_words_str += second_word
        all_remaining_words_list.extend(remaining)

    norm_ending = apply_char_map(all_ending_words_str)
    norm_second = apply_char_map(all_second_words_str)

    remaining_text_joined = "".join(all_remaining_words_list)
    norm_remaining_joined = apply_char_map(remaining_text_joined)

    unique_ending = sorted(set(norm_ending.replace(" ", "")))
    unique_second = sorted(set(norm_second.replace(" ", "")))

    print("--- 1. Ending Words ---")
    print(f"Unique Letters: {', '.join(unique_ending)}")
    print(f"Count: {len(unique_ending)}\n")

    print("--- 2. Second Words ---")
    print(f"Unique Letters: {', '.join(unique_second)}")
    print(f"Count: {len(unique_second)}\n")

    print("--- 3. Remaining Words ---")
    print(f"All Words: {' '.join(all_remaining_words_list)}")
    print(f"Total Letters in remaining words (after mapping): {len(norm_remaining_joined)}")

except Exception as e:
    print(f"Error occurred: {e}")
```

### **📖 The 'Inna' (إِنَّا) Pattern**

Surah Al-Kawthar begins with the word **إِنَّا** ('Inna'). The first verse in the Quran that starts with this word contains exactly **10 words**.

| Detail | Reference |
| :--- | :--- |
| **Surah** | Surah Al-Baqarah |
| **Verse** | Verse 119 |
| **Word Count** | 10 words |

---

```python
import pandas as pd


df = pd.read_excel(quran_url)


df.columns = ["VerseText", "Value"]


char_map = {'أ': 'ا', 'إ': 'ا', 'ء': 'ا', 'ئ': 'ا', 'ؤ': 'ا', 'آ': 'ا'}

def normalize(text):
    if not isinstance(text, str): return ""
    res = text
    for k, v in char_map.items():
        res = res.replace(k, v)
    return res


target_word = "انا"

first_match = None


for idx, row in df.iterrows():
    text = str(row["VerseText"]).strip()
    val = row["Value"]


    if val == 0 or not text:
        continue

    words = text.split()
    if words and normalize(words[0]) == target_word:
        first_match = text
        break


if first_match:
    word_list = first_match.split()
    count = len(word_list)
    print(f"First verse starting with 'Inna' (إنا):")
    print(f"Text: {first_match}")
    print(f"Total Words: {count}")

    if count == 10:
        print("Verification: SUCCESS. The verse contains exactly 10 words.")
    else:
        print(f"Verification: FAILED. The verse contains {count} words.")
else:
    print("No verse found starting with 'إنا'.")
```

```python
import pandas as pd

def get_digit_sum(n):
    """Calculates the sum of the digits of a number."""
    return sum(int(digit) for digit in str(n))

try:
    quran_df = pd.read_excel(quran_url)
    names_df = pd.read_csv(surah_names_url)
    abjad_df = pd.read_csv(abjad_url)
    print("Files loaded successfully.\n")
except Exception as e:
    print(f"Error loading files: {e}")
    exit()

quran_df.columns = ["VerseText", "Value"]
surah_arabic_names = names_df['Arabic'].tolist()
abjad_map = dict(zip(abjad_df['Letter'], abjad_df['Abjad Value']))

surahs_data = []
current_surah_verses = []

for index, row in quran_df.iterrows():
    text = str(row["VerseText"]).strip()
    value = row["Value"]

    if value == 0:
        if current_surah_verses:
            surahs_data.append(current_surah_verses)
            current_surah_verses = []
    else:
        if text and text != "nan":
            current_surah_verses.append(text)

if current_surah_verses:
    surahs_data.append(current_surah_verses)

alif_variations = ['أ', 'إ', 'آ', 'ء', 'ا', 'ئ', 'ؤ']
results = []

for idx, verses in enumerate(surahs_data):
    if len(verses) == 3:
        surah_name = surah_arabic_names[idx].strip()

        if surah_name[0] in alif_variations and surah_name[-1] == 'ر':
            surah_no = idx + 1
            full_text = "".join(verses).replace(" ", "")
            letter_count = len(full_text)

            if surah_no % 2 != 0 and letter_count % 2 != 0:
                parity_result = "Odd Parity"
            elif surah_no % 2 == 0 and letter_count % 2 == 0:
                parity_result = "Even Parity"
            else:
                parity_result = "Mixed (Non-Matching)"

            abjad_total = sum(abjad_map.get(char, 0) for char in full_text)

            results.append({
                "No": surah_no,
                "Name": surah_name,
                "Letters": letter_count,
                "Abjad": abjad_total,
                "Parity": parity_result,
                "AbjadDS": get_digit_sum(abjad_total)
            })

print(f"{'No':<5} | {'Name':<15} | {'Letters':<8} | {'Parity':<15} | {'Abjad':<10} | {'DS'}")
print("-" * 75)

for s in results:
    print(f"{s['No']:<5} | {s['Name']:<15} | {s['Letters']:<8} | {s['Parity']:<15} | {s['Abjad']:<10} | {s['AbjadDS']}")

if results:
    total_abjad_sum = sum(s["Abjad"] for s in results)
    final_ds_total = get_digit_sum(get_digit_sum(total_abjad_sum))

    sum_of_indiv_ds = sum(s["AbjadDS"] for s in results)
    final_ds_from_indiv = get_digit_sum(sum_of_indiv_ds)

    print("\n" + "="*60)
    print(f"Total Combined Abjad Value: {total_abjad_sum}")
    print(f"Final Recursive Digit Sum: {final_ds_total}")
    print("-" * 60)
    print(f"Sum of Individual Digit Sums: {sum_of_indiv_ds}")
    print(f"Final Digit Sum of Individual Sums: {final_ds_from_indiv}")
    print("="*60)
```

```python
import pandas as pd
import re

def calculate_abjad_from_file(csv_file_path):
    try:
        df = pd.read_csv(csv_file_path)

        abjad_values = dict(zip(df['Letter'], df['Abjad Value']))

    except FileNotFoundError:
        print(f"Error: File '{csv_file_path}' not found.")
        return
    except Exception as e:
        print(f"An error occurred reading the CSV: {e}")
        return

    def get_abjad_value(text):
        clean_text = re.sub(r'[\u064B-\u065F]', '', text)
        clean_text = clean_text.replace(" ", "")

        total = sum(abjad_values.get(char, 0) for char in clean_text)
        return total

    phrase_nineteen = "تسعة عشر"
    phrase_bismillah = "بسم الله الرحمن الرحيم"

    value_nineteen = get_abjad_value(phrase_nineteen)
    value_bismillah = get_abjad_value(phrase_bismillah)

    difference = value_nineteen - value_bismillah

    print("=" * 55)
    print("🔢 ABJAD VALUE ALIGNMENT (USING ATTACHED CSV)")
    print("=" * 55)

    print(f"Phrase 1: '{phrase_nineteen}' (Nineteen)")
    print(f"Total Abjad Value: {value_nineteen}")
    print("-" * 55)

    print(f"Phrase 2: '{phrase_bismillah}' (Bismillah)")
    print(f"Total Abjad Value: {value_bismillah}")
    print("-" * 55)

    print("MATHEMATICAL DIFFERENCE:")
    print(f"{value_nineteen} - {value_bismillah} = {difference}")
    print("=" * 55)


calculate_abjad_from_file(abjad_url)
```
