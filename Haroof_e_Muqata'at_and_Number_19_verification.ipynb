```python
def get_prime_order(n):
    if n < 2:
        return 0
    count = 0
    for i in range(2, n + 1):
        is_p = True
        for j in range(2, int(i**0.5) + 1):
            if i % j == 0:
                is_p = False
                break
        if is_p:
            count += 1
    return count

muqattaat_surah_numbers = [
    2, 3, 7, 10, 11, 12, 13, 14, 15, 19,
    20, 26, 27, 28, 29, 30, 31, 32, 36, 38,
    40, 41, 42, 43, 44, 45, 46, 50, 68
]

first_surah = min(muqattaat_surah_numbers)
last_surah = max(muqattaat_surah_numbers)

surahs_in_range = (last_surah - first_surah) + 1

is_prime = True
if surahs_in_range < 2:
    is_prime = False
else:
    for i in range(2, int(surahs_in_range**0.5) + 1):
        if surahs_in_range % i == 0:
            is_prime = False
            break

print("=== FIRST TO LAST MUQATTA'AT SURAH RANGE ===")
print(f"First Surah Number: {first_surah}")
print(f"Last Surah Number: {last_surah}")
print(f"Total Surahs from {first_surah} to {last_surah} (inclusive): {surahs_in_range}")
print("-" * 43)
print(f"Is {surahs_in_range} a Prime Number? {'Yes' if is_prime else 'No'}")

if is_prime:
    prime_order = get_prime_order(surahs_in_range)
    print(f"Order of Prime Number {surahs_in_range}: {prime_order}")
```

```python
import pandas as pd

# 1. File load karein
df = pd.read_excel(quran_url)

# 2. Surah boundaries identify karein (0 values as separators)
sep_indices = df[df['Numbers'] == 0].index.tolist()
sep_indices.append(len(df))

surah_counter = 0
start_idx = 0
tariq_verses = None

# 3. Surah 86 (Al-Tariq) ko dhoondein
for sep_idx in sep_indices:
    block = df.iloc[start_idx:sep_idx]
    verses = block[block['Verses'].notna()]

    if not verses.empty:
        surah_counter += 1
        if surah_counter == 86:
            tariq_verses = verses.copy()
            break
    start_idx = sep_idx + 1

if tariq_verses is not None:
    # Verse numbers assign karein (1, 2, 3...)
    tariq_verses['Verse_No'] = range(1, len(tariq_verses) + 1)

    # 4. Wo verses nikalein jin ki Abjad value 19 se divisible hai
    matches = tariq_verses[tariq_verses['Numbers'] % 19 == 0].copy()

    # 5. Verse numbers ka sum calculate karein
    v_sum = matches['Verse_No'].sum()

    # 6. Formatting aur Display
    print("-" * 60)
    print(f"{'Surah Al-Tariq (86) Analysis':^60}")
    print("-" * 60)
    print(f"{'V.No':<8} | {'Verse Text':<30} | {'Abjad':<10}")
    print("-" * 60)

    for _, row in matches.iterrows():
        print(f"{row['Verse_No']:<8} | {row['Verses']:<30} | {row['Numbers']:<10}")

    print("-" * 60)
    print(f"Total Matches: {len(matches)}")
    print(f"Sum of Verse Numbers: {matches['Verse_No'].tolist()} -> Sum = {v_sum}")
    print(f"Is Sum ({v_sum}) divisible by 19? {'YES (19 x ' + str(v_sum//19) + ')' if v_sum % 19 == 0 else 'NO'}")
    print("-" * 60)

else:
    print("Error: Surah 86 file mein nahi mili.")
```
