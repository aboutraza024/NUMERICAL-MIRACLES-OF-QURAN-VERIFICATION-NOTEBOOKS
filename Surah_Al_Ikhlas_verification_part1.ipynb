```python
def get_prime_factors(n):
    i, factors = 2, []
    temp = n
    while i * i <= temp:
        if temp % i: i += 1
        else:
            temp //= i
            factors.append(i)
    if temp > 1: factors.append(temp)
    return factors

def get_prime_order(n):
    if n < 2: return None
    primes = []
    c = 2
    while True:
        is_p = all(c % p != 0 for p in primes)
        if is_p:
            primes.append(c)
            if c == n:
                return len(primes)
        if c > n:
            break
        c += 1
    return None

def get_abjad_map():
    return {
        'ا': 1, 'ب': 2, 'ج': 3, 'د': 4, 'ه': 5, 'و': 6, 'ز': 7, 'ح': 8, 'ط': 9,
        'ي': 10, 'ك': 20, 'ل': 30, 'م': 40, 'ن': 50, 'س': 60, 'ع': 70, 'ف': 80,
        'ص': 90, 'ق': 100, 'ر': 200, 'ش': 300, 'ت': 400, 'ث': 500, 'خ': 600,
        'ذ': 700, 'ض': 800, 'ظ': 900, 'غ': 1000
    }

def analyze_quranic_logic(surah, verse_no, text):
    total_sv_sum = surah + verse_no
    sv_factors = get_prime_factors(total_sv_sum)
    biggest_pf_sv = max(sv_factors) if sv_factors else 0
    digit_sum_sv = sum(int(d) for d in str(total_sv_sum))

    normalization = {
        'أ': 'ا', 'إ': 'ا', 'آ': 'ا', 'ء': 'ا', 'ئ': 'ا', 'ؤ': 'ا',
        'ة': 'ه', 'ى': 'ي'
    }
    abjad_map = get_abjad_map()
    clean_text = "".join(text.split())

    normalized_list = []
    for char in clean_text:
        normalized_list.append(normalization.get(char, char))

    unique_letters = sorted(list(set(normalized_list)))
    unique_abjad_sum = sum(abjad_map.get(char, 0) for char in unique_letters)

    unique_factors = get_prime_factors(unique_abjad_sum)
    biggest_pf_unique = max(unique_factors) if unique_factors else 0
    digit_sum_unique_pf = sum(int(d) for d in str(biggest_pf_unique))

    p_order = get_prime_order(digit_sum_unique_pf)

    print(f"--- Surah {surah}, Verse {verse_no} Analysis ---")
    print(f"Sum (Surah+Verse): {total_sv_sum}")
    print(f"Biggest Prime Factor of Sum: {biggest_pf_sv}")
    print(f"Digit Sum of Sum (98): {digit_sum_sv}")

    print("\n" + "-"*40 + "\n")

    print(f"Verse: {text}")
    print(f"Unique Letters Count: {len(unique_letters)}")
    print(f"Abjad Sum of Unique Letters: {unique_abjad_sum}")
    print(f"Biggest Prime Factor of Abjad Sum: {biggest_pf_unique}")
    print(f"Digit Sum of {biggest_pf_unique}: {digit_sum_unique_pf}")

    if p_order:
        print(f"Prime Order of {digit_sum_unique_pf}: {p_order}")
    else:
        print(f"The number {digit_sum_unique_pf} is not a prime number.")

surah_num = 71
verse_num = 27
verse_txt = "انك إن تذرهم يضلوا عبادك ولا يلدوا إلا فاجرا كفارا"

analyze_quranic_logic(surah_num, verse_num, verse_txt)
```

```python
import pandas as pd

def get_nth_prime(n):
    primes = []
    num = 2
    while len(primes) < n:
        for p in primes:
            if num % p == 0:
                break
        else:
            primes.append(num)
        num += 1
    return primes[-1]

def get_digit_sum(number):
    return sum(int(digit) for digit in str(number))

df = pd.read_excel(quran_url)

mask = (df['Numbers'] != 0) | (df['Numbers'].diff() != 0)
df_clean = df[mask].copy()
df_clean['Surah_Number'] = (df_clean['Numbers'] == 0).cumsum() + 1

verses_df = df_clean[df_clean['Numbers'] != 0].copy()
verses_df['Verse_Number'] = verses_df.groupby('Surah_Number').cumcount() + 1

target_surah = 39
target_verse = 6

surah_data = verses_df[verses_df['Surah_Number'] == target_surah]
verse_data = surah_data[surah_data['Verse_Number'] == target_verse]

if not verse_data.empty:
    verse_text = verse_data['Verses'].values[0]
    print(f"Verse Text (39:6): {verse_text}\n")

surah_digits = [int(d) for d in str(target_surah)]
verse_digits = [int(d) for d in str(target_verse)]

all_digits = surah_digits + verse_digits

sum_of_digits = sum(all_digits)

final_digit_sum = get_digit_sum(sum_of_digits)

nth_prime = get_nth_prime(final_digit_sum)

print(f"Step 1: The numbers extracted from 39:6 are -> {all_digits[0]}, {all_digits[1]}, and {all_digits[2]}")
print(f"Step 2: Sum of these numbers ({all_digits[0]} + {all_digits[1]} + {all_digits[2]}) -> {sum_of_digits}")
print(f"Step 3: Digit sum of {sum_of_digits} -> {final_digit_sum}")
print(f"Step 4: The {final_digit_sum}th Prime Number is -> {nth_prime}")
```
