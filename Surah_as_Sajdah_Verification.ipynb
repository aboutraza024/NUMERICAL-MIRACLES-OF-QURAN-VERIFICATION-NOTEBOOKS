## **3. Prime Factor Analysis of 186**
Breaking down the central word's position (186) reveals a hidden link to the total verse count.

### **The Prime Breakdown:**
$$186 = 2 \times 3 \times \mathbf{31}$$

### **The Prime Order Chain:**
1.  **31** is the **11th** prime number.
2.  **11** is the **5th** prime number.

### **The Mathematical Substitution:**
If we replace the large factor (**31**) with its prime order equivalent (**5**):
# <center> $2 \times 3 \times \mathbf{5} = \mathbf{30}$ </center>

> **Result:** This matches the **Total Verses** in Surah As-Sajdah exactly.

---


```python
import pandas as pd

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def get_prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

def get_prime_order(prime_num):
    if not is_prime(prime_num):
        return None
    count = 0
    num = 2
    while True:
        if is_prime(num):
            count += 1
            if num == prime_num:
                return count
        num += 1

df = pd.read_excel(quran_url)
```

```python
target_word = "سجدا"

print("=" * 50)

if target_word in verse_words:
    word_index = verse_words.index(target_word)

    verse_word_position = word_index + 1
    words_before = word_index
    words_after = len(verse_words) - (word_index + 1)
    total_words_verse = len(verse_words)

    print("Verse Text:", verse_text)
    print("-" * 50)
    print(f"Total words in the verse: {total_words_verse}")
    print(f"Position of the word '{target_word}': {verse_word_position}")
    print(f"Number of words BEFORE '{target_word}': {words_before}")
    print(f"Number of words AFTER '{target_word}': {words_after}")
    print("-" * 50)
    print("Sum Breakdown:")
    print(f"{words_before} + 1 + {words_after} = {total_words_verse}")

else:
    print(f"The word '{target_word}' was not found in Verse 15.")

print("\n" + "=" * 50)

if target_word in all_words_in_surah:
    surah_word_position = all_words_in_surah.index(target_word) + 1
    total_words_surah = len(all_words_in_surah)
    total_verses = len(surah_32_verses)

    print(f"Overall position of '{target_word}' in Surah As-Sajdah: {surah_word_position}")
    print(f"Total words in Surah As-Sajdah: {total_words_surah}")
    print(f"Total verses in Surah As-Sajdah: {total_verses}")

    if surah_word_position * 2 == total_words_surah:
        print(f"✔ Verification: {surah_word_position} * 2 = {total_words_surah}")

    print("\n--- Prime Factor & Order Calculation ---")

    prime_factors = get_prime_factors(surah_word_position)
    print(f"Prime factors of {surah_word_position}: {prime_factors}")

    highest_prime = max(prime_factors)
    print(f"Highest prime factor: {highest_prime}")

    order_of_highest = get_prime_order(highest_prime)
    print(f"Order of {highest_prime} in prime sequence: {order_of_highest}")

    if order_of_highest is not None and is_prime(order_of_highest):
        order_position = get_prime_order(order_of_highest)

        print(f"The order '{order_of_highest}' is also prime. Its position is: {order_position}")

        new_factors = prime_factors.copy()
        highest_index = new_factors.index(highest_prime)
        new_factors[highest_index] = order_position

        final_product = 1
        for f in new_factors:
            final_product *= f

        print(f"Replaced factors: {new_factors}")
        print(f"Final multiplied result: {final_product}")

    else:
        print(f"The order {order_of_highest} is not prime or invalid.")

else:
    print(f"The word '{target_word}' was not found in Surah As-Sajdah.")
```

# <center> **131** $-$ **113** $=$ **18** </center>

```python

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def get_nth_prime(n):
    count = 0
    num = 2
    while True:
        if is_prime(num):
            count += 1
            if count == n:
                return num
        num += 1

chapter_number = 32
verses_count = 30

prime_32 = get_nth_prime(chapter_number)
prime_30 = get_nth_prime(verses_count)

difference = prime_32 - prime_30
print(f"Surah Sajdah Chapter Number: {chapter_number}")
print(f"Surah Sajdah Verses Count: {verses_count}\n")

print(f"The {chapter_number}nd Prime Number is: {prime_32}")
print(f"The {verses_count}th Prime Number is: {prime_30}\n")

print(f"Difference: {prime_32} - {prime_30} = {difference}")
```
