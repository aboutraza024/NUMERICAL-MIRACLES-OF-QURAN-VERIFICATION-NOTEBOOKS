```python


import pandas as pd

TARGET_SURAH = 16
TARGET_AYAH = 16

df = pd.read_excel(quran_url)

surah_nums = []
ayah_nums = []

current_surah = 1
current_ayah = 0
prev_blank = False

for verse in df["Verses"]:
    if isinstance(verse, str) and verse.strip():
        current_ayah += 1
        surah_nums.append(current_surah)
        ayah_nums.append(current_ayah)
        prev_blank = False
    else:
        surah_nums.append(None)
        ayah_nums.append(None)
        if not prev_blank:
            current_surah += 1
            current_ayah = 0
        prev_blank = True

df["Surah"] = surah_nums
df["Ayah"] = ayah_nums


ARABIC_ALPHABET = list("ابتثجحخدذرزسشصضطظعغفقكلمنهوي")

VARIANT_MAP = {
    "أ": "ا", "إ": "ا", "آ": "ا", "ٱ": "ا", "ئ": "ا", "ؤ": "ا", "ء": "ا",
    "ة": "ه",
    "ى": "ي",
}


def normalize_letter(ch: str):
    """
    Normalize a character to one of the 28 base Arabic letters:
    - Map common variants (أ، إ، آ، ٱ، ؤ، ئ، ء -> ا; ة -> ه; ى -> ي)
    - Ignore anything that is not a letter (spaces, punctuation, etc.)
    """
    if ch in VARIANT_MAP:
        return VARIANT_MAP[ch]
    if ch in ARABIC_ALPHABET:
        return ch
    return None


row = df[(df["Surah"] == TARGET_SURAH) & (df["Ayah"] == TARGET_AYAH)].iloc[0]
verse_text = row["Verses"]

present_letters = set()

for ch in verse_text:
    base = normalize_letter(ch)
    if base is not None:
        present_letters.add(base)

present_in_order = [l for l in ARABIC_ALPHABET if l in present_letters]
absent_letters = [l for l in ARABIC_ALPHABET if l not in present_letters]

print(f"Surah {TARGET_SURAH}, Ayah {TARGET_AYAH} text:")
print(verse_text)
print("")



print("\nAbsent letters (from 28 hijai letters):")
print(" ".join(absent_letters))
print(f"Number of absent letters: {len(absent_letters)}")
```

# Arabic Letters and Their Variations  
*Variations are **not counted as separate letters**; they are treated as their base letter.*

- **ا (Alif)**  
  Variations: أ, إ, آ, ء, ؤ, ئ  
  _All these are counted as "ا"._

- **ي (Ya)**  
  Variations: ى  
  _Both ي and ى are counted as "ي"._

- **ه (Ha)**  
  Variations: ة  
  _Both ه and ة are counted as "ه"._

```python
import pandas as pd


alif_variations = ['ا', 'أ', 'إ', 'آ', 'ء', 'ؤ', 'ئ']
ya_variations = ['ي', 'ى']
ha_variations = ['ه', 'ة']


dataset = pd.read_excel(quran_url)

target_surah, target_verse = 16, 68
current_surah, current_verse = 1, 1
verse_text = ""

for entry in dataset['Verses']:
    if pd.isna(entry) or str(entry).strip() == "":
        current_surah += 1
        current_verse = 1
        continue
    if current_surah == target_surah and current_verse == target_verse:
        verse_text = str(entry).strip()
        break
    current_verse += 1


letters_only = [char for char in verse_text if char != ' ']
unique_base_letters = set()

for char in letters_only:
    if char in alif_variations:
        unique_base_letters.add('BASE_ALIF')
    elif char in ya_variations:
        unique_base_letters.add('BASE_YA')
    elif char in ha_variations:
        unique_base_letters.add('BASE_HA')
    else:
        unique_base_letters.add(char)


print(f"Total Letters: {len(letters_only)}")
print(f"Unique Base Letters: {len(unique_base_letters)}")
print(f"Unique Letters List: {sorted(list(unique_base_letters))}")
```

```python

import pandas as pd
import re
import unicodedata


TARGET_SURAH_NAHL = 16
TARGET_WORD_RAW = "النحل"


df = pd.read_excel(quran_url)


surah_nums = []
ayah_nums = []

current_surah = 1
current_ayah = 0
prev_blank = False

for verse in df["Verses"]:
    if isinstance(verse, str) and verse.strip():
        current_ayah += 1
        surah_nums.append(current_surah)
        ayah_nums.append(current_ayah)
        prev_blank = False
    else:
        surah_nums.append(None)
        ayah_nums.append(None)
        if not prev_blank:
            current_surah += 1
            current_ayah = 0
        prev_blank = True

df["Surah"] = surah_nums
df["Ayah"] = ayah_nums


def count_words(text: str) -> int:
    """Simple word count: split by whitespace."""
    if not isinstance(text, str):
        return 0
    return len(text.split())


ARABIC_LETTER_RE = re.compile(r"[\u0621-\u064A]")

def count_letters(text: str) -> int:
    if not isinstance(text, str):
        return 0
    return sum(1 for ch in text if ARABIC_LETTER_RE.match(ch))


def strip_diacritics(s: str) -> str:
    """Remove Arabic diacritics (tashkeel) using Unicode combining marks."""
    decomposed = unicodedata.normalize("NFD", s)
    without_marks = "".join(
        ch for ch in decomposed if unicodedata.category(ch) != "Mn"
    )
    return unicodedata.normalize("NFC", without_marks)


def normalize_arabic_word(w: str) -> str:
    """
    Normalize an Arabic word for comparison:
    - Remove diacritics
    - Unify بعض أشكال الألف (ٱ، أ، إ، آ -> ا)
    - Remove non-Arabic characters (punctuation, digits, etc.)
    """
    w = strip_diacritics(w)
    w = (w.replace("ٱ", "ا")
           .replace("أ", "ا")
           .replace("إ", "ا")
           .replace("آ", "ا"))
    w = re.sub(r"[^\u0600-\u06FF]", "", w)
    return w


TARGET_WORD_NORM = normalize_arabic_word(TARGET_WORD_RAW)


ayah16_df = df[(df["Ayah"] == 16) & df["Verses"].notna()].copy()

ayah16_df["word_count"] = ayah16_df["Verses"].apply(count_words)
ayah16_df["letter_count"] = ayah16_df["Verses"].apply(count_letters)

num_surahs_with_16 = ayah16_df["Surah"].nunique()
total_words_in_ayah16 = int(ayah16_df["word_count"].sum())
total_letters_in_ayah16 = int(ayah16_df["letter_count"].sum())

print("=== A) Verse 16 across the Quran ===")
print(f"Number of surahs that have Ayah 16            : {num_surahs_with_16}")
print(f"Total number of words in ALL Ayah 16         : {total_words_in_ayah16}")
print(f"Total number of letters in ALL Ayah 16       : {total_letters_in_ayah16}")
print()



nahl_1_68 = df[
    (df["Surah"] == TARGET_SURAH_NAHL)
    & df["Ayah"].notna()
    & (df["Ayah"] >= 1)
    & (df["Ayah"] <= 68)
].copy()

nahl_1_68["word_count"] = nahl_1_68["Verses"].apply(count_words)
nahl_1_68["letter_count"] = nahl_1_68["Verses"].apply(count_letters)

total_words_nahl_1_68 = int(nahl_1_68["word_count"].sum())
total_letters_nahl_1_68 = int(nahl_1_68["letter_count"].sum())

print("=== B) Surah an-Nahl, verses 1–68 ===")

print(f"Total letters from start of Surah an-Nahl to Ayah 68 : {total_letters_nahl_1_68}")
print()



nahl_all = df[
    (df["Surah"] == TARGET_SURAH_NAHL) & df["Verses"].notna()
].copy()

total_words_until_target = 0
found = False
found_ayah = None

for _, row in nahl_all.iterrows():
    ayah_no = int(row["Ayah"])
    verse_text = str(row["Verses"])
    words = verse_text.split()

    for w in words:
        total_words_until_target += 1
        if normalize_arabic_word(w) == TARGET_WORD_NORM:
            found = True
            found_ayah = ayah_no
            break

    if found:
        break

print("=== C) Words in Surah an-Nahl until the word 'النحل' appears ===")
if found:
    print(f"'النحل' found in Surah {TARGET_SURAH_NAHL}, Ayah {found_ayah}.")
    print(f"Total number of words from the start of Surah an-Nahl up to and including the word 'النحل': {total_words_until_target}")
else:
    print("The word 'النحل' was NOT found in Surah an-Nahl using current normalization.")
```

```python
import pandas as pd

def get_prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors


dataset = pd.read_excel(quran_url)

surahs_metadata = []
current_surah_verses = []
surah_count = 0

for entry in dataset['Verses']:
    if pd.isna(entry) or str(entry).strip() == "":
        if current_surah_verses:
            surah_count += 1

            word_count = sum(len(str(v).split()) for v in current_surah_verses)
            surahs_metadata.append({
                'surah_no': surah_count,
                'verse_count': len(current_surah_verses),
                'word_count': word_count
            })
            current_surah_verses = []
        if surah_count == 16:
            break
        continue
    current_surah_verses.append(entry)

surahs_with_v68 = [s for s in surahs_metadata if s['verse_count'] >= 68]
count_v68 = len(surahs_with_v68)

surah_13 = surahs_metadata[12]
s13_verses = surah_13['verse_count']
s13_words = surah_13['word_count']


difference = s13_verses - 13
final_sum = s13_words + difference
factors = get_prime_factors(final_sum)


print(f"--- Global Quranic Pattern Verification ---")
print(f"1. Surahs (1-16) containing Verse 68: {count_v68}")
print(f"2. Surah 13 (Ar-Ra'd) Verse Count: {s13_verses}")
print(f"3. Surah 13 (Ar-Ra'd) Word Count: {s13_words}")
print(f"4. Surah 13 Difference (Verse Count - Surah No): {s13_verses} - 13 = {difference}")
print(f"5. Final Sum (Words + Difference): {s13_words} + {difference} = {final_sum}")
print(f"6. Prime Factors of {final_sum}: {factors}")

if len(factors) >= 3:
    middle_idx = len(factors) // 2
    print(f"   Middle Factor: {factors[middle_idx]}")
```

### **Arabic word from "honey" (عسل) 🍯 & its Arabic letter positions**

* **ع (Ayn)** = 18th letter
* **س (Seen)** = 12th letter
* **ل (Laam)** = 23rd letter

> **Sum = 53 ➔ 16th prime number**

```python


arabic_alphabet = list("ابتثجحخدذرزسشصضطظعغفقكلمنهوي")

def letter_order(letter: str) -> int:
    """
    Return 1-based order of the letter in the hijai alphabet.
    Raises ValueError if letter not in the list.
    """
    return arabic_alphabet.index(letter) + 1

def is_prime(n: int) -> bool:
    """Check if n is a prime number."""
    if n < 2:
        return False
    if n in (2, 3):
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

def prime_index(p: int):
    """
    If p is prime, return its index in the sequence of primes:
    2 -> 1, 3 -> 2, 5 -> 3, ...
    If p is not prime, return None.
    """
    if not is_prime(p):
        return None

    count = 0
    num = 2
    while num <= p:
        if is_prime(num):
            count += 1
            if num == p:
                return count
        num += 1
    return None

word = "عسل"


orders = [letter_order(ch) for ch in word]


sum_orders = sum(orders)

is_sum_prime = is_prime(sum_orders)
prime_pos = prime_index(sum_orders) if is_sum_prime else None

print(f"Word          : {word}")
print(f"Letter orders : {orders}")
print(f"Sum of orders : {sum_orders}")
print(f"Is sum prime? : {is_sum_prime}")
if is_sum_prime:
    print(f"Prime index   : {prime_pos}  (position of {sum_orders} among the primes)")
```
