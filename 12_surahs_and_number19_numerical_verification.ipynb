### Parity Pattern
> **Note:** Among these 12 surahs, 6 have matching parity (both order and verses are odd or both even), and 6 have non-matching parity (one odd, one even).

```python

import pandas as pd


df = pd.read_excel(quran_url)


surah_lengths = []
current_length = 0

for index, row in df.iterrows():
    if row['Numbers'] == 0:
        if current_length > 0:
            surah_lengths.append(current_length)
            current_length = 0
    else:
        current_length += 1

if current_length > 0:
    surah_lengths.append(current_length)

print(f"Total Number of Surahs extracted: {len(surah_lengths)}")

matching_group = []
non_matching_group = []

for i, length in enumerate(surah_lengths):
    surah_number = i + 1
    total_sum = surah_number + length

    if total_sum % 19 == 0:
        factor = total_sum // 19
        factor_equation = f"{total_sum} = 19 * {factor}"

        surah_is_even = (surah_number % 2 == 0)
        verses_is_even = (length % 2 == 0)

        item = {
            'Surah Number': surah_number,
            'Verses': length,
            'Sum': total_sum,
            'Factor': factor,
            'Parity': '',
            'Equation': factor_equation
        }

        if surah_is_even and verses_is_even:
            item['Parity'] = "Even-Even"
            matching_group.append(item)
        elif not surah_is_even and not verses_is_even:
            item['Parity'] = "Odd-Odd"
            matching_group.append(item)
        elif surah_is_even and not verses_is_even:
            item['Parity'] = "Even-Odd"
            non_matching_group.append(item)
        else:
            item['Parity'] = "Odd-Even"
            non_matching_group.append(item)

print(f"Total Surahs divisible by 19: {len(matching_group) + len(non_matching_group)}")
```

```python

def calculate_group_stats(group):
    total_factors_sum = sum(item['Factor'] for item in group)
    unique_sums = sorted(list(set(item['Sum'] for item in group)), reverse=True)

    sum_difference = unique_sums[0] - unique_sums[1] if len(unique_sums) >= 2 else unique_sums[0]

    sum_surah_numbers = sum(item['Surah Number'] for item in group)
    sum_verses = sum(item['Verses'] for item in group)
    total_group_sum = sum_surah_numbers + sum_verses
    is_divisible_19 = (total_group_sum % 19 == 0)
    group_factor = total_group_sum // 19

    return {
        'factor_sum': total_factors_sum,
        'diff': sum_difference,
        'grand_total': total_group_sum,
        'is_div19': is_divisible_19,
        'group_factor': group_factor
    }

g1 = calculate_group_stats(matching_group)
g2 = calculate_group_stats(non_matching_group)

odd_odd_sum = sum(item['Sum'] for item in matching_group if item['Parity'] == 'Odd-Odd')
even_even_sum = sum(item['Sum'] for item in matching_group if item['Parity'] == 'Even-Even')
even_odd_sum = sum(item['Sum'] for item in non_matching_group if item['Parity'] == 'Even-Odd')
odd_even_sum = sum(item['Sum'] for item in non_matching_group if item['Parity'] == 'Odd-Even')

print("Calculations Ready for Display!")
print("========== GROUP 1: MATCHING PARITY ==========\n")
display(pd.DataFrame(matching_group).drop(columns=['Factor']))

print(f"\n--- Sub-Groups inside Matching Parity ---")
print(f"Odd-Odd Group (3 Surahs): Total Sum = {odd_odd_sum} (19 * {odd_odd_sum//19})")
print(f"Even-Even Group (3 Surahs): Total Sum = {even_even_sum} (19 * {even_even_sum//19})")

print(f"\n--- Group 1 Calculations ---")
print(f"Difference b/w Highest & 2nd Highest Sums = {g1['diff']}")
print(f"Sum of 19's Multipliers = {g1['factor_sum']}")
print(f"Grand Total (Surahs + Verses) = {g1['grand_total']} (19 * {g1['group_factor']})\n")
```
