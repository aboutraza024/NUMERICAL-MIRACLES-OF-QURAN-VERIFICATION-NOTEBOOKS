## **⚖️ The Perfect Balance**
**A remarkable symmetry centered around the letter Ta (ط)**

---

### **🟢 The Central Letter**
The center **Ta (ط)** of the entire Quran is found in the first verse of **Surah Ta-Ha (طه)**.

* **Center Position:** 637th **ط** in the Quran.
* **Huroof-e-Muqatta'at:** This verse contains only two letters (**ط** and **ه**).

---

### **📊 Symmetry Analysis**
The distribution of the letter **Ta (ط)** across the Quran is perfectly balanced around this center point:

| Position | Count of Letters **ط** |
| :--- | :---: |
| Letters **ط** on the **LEFT** (Before Center) | **636** |
| **Center Letter (ط)** | **637th** |
| Letters **ط** on the **RIGHT** (After Center) | **636** |

---

> **Summary:**
> **636** (Before) + **1** (Center) + **636** (After) = **1273** total occurrences of the letter **Ta**.
---

```python
import pandas as pd


df = pd.read_excel(quran_url)

occurrences = []
current_surah = 1
current_verse = 1
last_was_empty = False

for idx, row in df.iterrows():
    verse_text = row['Verses']


    if pd.isna(verse_text) or str(verse_text).strip() == "":
        if not last_was_empty:
            current_surah += 1
            current_verse = 1
            last_was_empty = True
        continue

    last_was_empty = False
    verse_str = str(verse_text)


    for pos, char in enumerate(verse_str):
        if char == 'ط':
            occurrences.append({
                'surah': current_surah,
                'verse': current_verse,
                'text': verse_str
            })

    current_verse += 1


total_tah = len(occurrences)
mid_idx = total_tah // 2
central = occurrences[mid_idx]


print(f"Total 'ط' Count: {total_tah}")
print(f"Central 'ط' Sequence: {mid_idx + 1}")
print(f"Before: {mid_idx} | After: {total_tah - mid_idx - 1}")
print(f"Surah Number: {central['surah']}")
print(f"Verse Number: {central['verse']}")
print(f"Verse Text: {central['text']}")
```

```python
import pandas as pd

dataset = pd.read_excel(quran_url)

target_letter = 'ط'
max_frequency = 0
results = []

current_surah = 1
current_verse = 1
surah_transition = False

for entry in dataset['Verses']:
    if pd.isna(entry) or str(entry).strip() == "":
        if not surah_transition:
            current_surah += 1
            current_verse = 1
            surah_transition = True
        continue
    surah_transition = False
    count = str(entry).count(target_letter)
    if count > max_frequency:
        max_frequency = count
    current_verse += 1

current_surah = 1
current_verse = 1
surah_transition = False

for entry in dataset['Verses']:
    if pd.isna(entry) or str(entry).strip() == "":
        if not surah_transition:
            current_surah += 1
            current_verse = 1
            surah_transition = True
        continue
    surah_transition = False
    if str(entry).count(target_letter) == max_frequency:
        results.append({
            'surah': current_surah,
            'verse': current_verse,
            'text': str(entry)
        })
    current_verse += 1

surah_sum = sum(ref['surah'] for ref in results)
verse_sum = sum(ref['verse'] for ref in results)
grand_total = surah_sum + verse_sum

print(f"Maximum occurrences of '{target_letter}' in a single verse: {max_frequency}")
print(f"Total number of verses containing the maximum frequency: {len(results)}\n")

for idx, ref in enumerate(results, 1):
    print(f"Reference {idx}: Surah {ref['surah']}, Verse {ref['verse']}")
    print(f"Text: {ref['text']}\n")

print(f"Sum of Surah Numbers: {surah_sum}")
print(f"Sum of Verse Numbers: {verse_sum}")
print(f"Total Combined Sum (Surahs + Verses): {grand_total}")
```

How many prime numbers occur before 58

```python
def check_prime_status(integer):
    if integer < 2:
        return False
    for factor in range(2, int(integer**0.5) + 1):
        if integer % factor == 0:
            return False
    return True

upper_limit = 58
prime_sequence = [num for num in range(2, upper_limit) if check_prime_status(num)]
total_primes = len(prime_sequence)

print(f"Prime numbers before {upper_limit}: {prime_sequence}")
print(f"Total count: {total_primes}")
```

How many Additive Prime Numbers before 116

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

def get_sum_of_digits(number):
    return sum(int(digit) for digit in str(number))

upper_limit = 116
additive_primes = []

for num in range(2, upper_limit):
    if is_prime(num):
        digit_sum = get_sum_of_digits(num)
        if is_prime(digit_sum):
            additive_primes.append(num)

total_count = len(additive_primes)

print(f"Additive Prime Numbers before {upper_limit}: {additive_primes}")
print(f"Total Count: {total_count}")
```

**How many Additive Composites before 116**

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

def is_composite(number):
    return number > 1 and not is_prime(number)

def get_sum_of_digits(number):
    return sum(int(digit) for digit in str(number))

def get_prime_index(target):
    count = 0
    num = 2
    while True:
        if is_prime(num):
            count += 1
            if num == target:
                return count
        num += 1

upper_limit = 116
additive_composites = []

for num in range(2, upper_limit):
    if is_composite(num):
        digit_sum = get_sum_of_digits(num)
        if is_composite(digit_sum):
            additive_composites.append(num)

total_count = len(additive_composites)

print(f"Additive Composite Numbers before {upper_limit}:")
print(additive_composites)
print(f"\nTotal Count: {total_count}")

if is_prime(total_count):
    prime_order = get_prime_index(total_count)
    print(f"✅ Total count {total_count} is a PRIME number.")
    print(f"🔹 It is the {prime_order}th prime number.")
else:
    print(f"❌ Total count {total_count} is NOT a prime number.")
```

```python
import pandas as pd

def is_composite(number):
    if number <= 3:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return True
    return False

def get_nth_composite(n):
    count = 0
    candidate = 4
    while True:
        if is_composite(candidate):
            count += 1
            if count == n:
                return candidate
        candidate += 1


dataset = pd.read_csv(abjad_url)
target_letter = 'ط'
letter_row = dataset[dataset['Letter'] == target_letter]

if not letter_row.empty:
    abjadi_order = letter_row.index[0] + 1
    abjad_value = letter_row.iloc[0]['Abjad Value']


    composite_result = get_nth_composite(abjadi_order)

    print(f"Letter: {target_letter}")
    print(f"Abjadi Order: {abjadi_order}")
    print(f"Abjad Value: {abjad_value}")
    print(f"The {abjadi_order}th Composite Number: {composite_result}")
```
